## 2 - Read the data with erros again and insert the median of each attribute in place of the NaNs

```python
# Reading data from iris w/erros again and replacing '?' chars
data = pd.read_csv('data/iris-with-errors.csv', header=(0))
data = data.replace('?', np.nan)
data = data.drop_duplicates()

print("Shape:", data.shape)
data.head(20)
```

```python
# Now we get the median of each attribute

# First convert to a numpy array (except the last column which is qualitative)
data_arr = np.array(data.drop({'species'}, axis=1), np.float32)

# Then we get the median of each column
col_number = data_arr.shape[1]
medians = [np.nanmedian(data_arr[:, i]) for i in range(col_number)]

# A last we replace the NaNs from the dataframe with the medians
for i in range(col_number):
    data[data.columns[i]] = data[data.columns[i]].replace(np.nan, medians[i])

data.head(20)
```

## 3 - Consider the data from Iris and show the probability distributions of each of the variables after the normalization and standardization

```python
from sklearn.preprocessing import MinMaxScaler

data = pd.read_csv('data/iris.csv', header=(0))
data.head(20)
```

```python
from sklearn.preprocessing import MinMaxScaler
# Getting the cols containing numerical data
data_arr = np.array(data.drop("species", axis=1))

# Normalizing data
scaler = MinMaxScaler(feature_range=(0, 1))
data_arr_norm = scaler.fit_transform(data_arr)

print(data_arr_norm)
```

```python
print("Max of each attribute:", np.amax(data_arr_norm, axis=0))
print("Min of each attribute:", np.amin(data_arr_norm, axis=0))
```

```python
# Now lets plot the histogram for each of the features
for i in range(data_arr_norm.shape[1]):
    plt.hist(data_arr_norm[:,i])
    plt.title(data.columns[i])
    plt.show()
```

```python
# Now let's do the standardization
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler().fit(data_arr)
data_arr_rescaled = scaler.transform(data_arr)
print(data_arr_rescaled)
```

```python
# We can test our standardization by looking at the mean and std for each attribute
for col in data_arr_rescaled.T:
    print("Mean:", np.mean(col))
    print("Std:", np.std(col), "\n")
```

```python
# Plot the histograms again
for i in range(data_arr_rescaled.shape[1]):
    plt.hist(data_arr_rescaled[:,i])
    plt.title(data.columns[i])
    plt.show()
```

## 6 - Find the outliers in Iris and BostonHouse data

```python
# Lets start by importing the data from Iris again
data = pd.read_csv("data/iris.csv", header=(0))

# We do the analysis for each class separately

# First we divide the classes
labels = data["species"].unique()
class_data_v = [data.query('species == "{}"'.format(label)) for label in labels]
new_data = pd.DataFrame()

# Now finding the outliers for each class and checking
for class_data in class_data_v:
    Q1 = class_data.quantile(0.25)
    Q3 = class_data.quantile(0.75)
    IQR = Q3 - Q1
    
    is_outlier = ((class_data < (Q1 - 1.5 * IQR)) | (class_data > (Q3 + 1.5 * IQR))).any(axis=1)
    
    # Removing outliers
    class_data = class_data.drop(class_data.index[is_outlier], axis=0)
    new_data = pd.concat([new_data,class_data])

sns.pairplot(new_data, hue='species', corner=True)
plt.show()
```

```python
# We can do a new boxplot to check
labels = data["species"].unique()
for label in labels:
    query_expr = "species == '{}'".format(label)
    class_data = data.query(query_expr)
    
    sns.boxplot(data=class_data)
    plt.title(label)
    plt.ylim(-1, 10)
    plt.show()
```

```python
# Now repeating the process for the BostonHouses data
data = pd.read_csv("data/BostonHousing.csv", header=(0))

# First lets take a look at the data
print("Shape:", data.shape)
data
```

```python
# And get the correlation in our data
corr = data.corr()
# Displaying the matrix
plt.figure(figsize=(7, 7))
plt.imshow(corr, cmap='Blues', interpolation='none', aspect='auto')
plt.colorbar()
plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.yticks(range(len(corr)), corr.columns);
plt.suptitle('Correlation between variables', fontsize=15, fontweight='bold')
plt.grid(False)
plt.show()
```

```python
# Unlike for Iris, we don't have separate classes, so we just remove the outliers of the whole dataset

Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1

is_outlier = ((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)
print("Number of outliers found:", np.sum(is_outlier))

new_data = data.drop(data.index[is_outlier], axis=0)
```

```python
sns.boxplot(data=data)
plt.title("Before cleaning")
plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.show()

sns.boxplot(data=new_data)
plt.title("After cleaning")
plt.xticks(range(len(corr)), corr.columns, rotation='vertical')
plt.show()
```

## 8 - Do a PCA on the data from BostonHouse

```python
# Loading data
data = pd.read_csv("data/BostonHousing.csv", header=(0))
print(data.shape)
data.head(20)
```

```python
# We start by doing a standarization on our data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler().fit(data)
scaled_data = scaler.transform(data)
```

```python
# Now we do a Principal Component Analysis
from sklearn.decomposition import PCA

# Selecting 2 principal components
pca = PCA(n_components=2)
pca_result = pca.fit_transform(scaled_data)
```

```python
# Ploring our results...
plt.figure(figsize=(8,5))
plt.scatter(pca_result[:,0], pca_result[:,1], s=50, color = 'blue')
plt.xlabel("First component", fontsize=20)
plt.ylabel("Second component", fontsize=20)
plt.xticks(color='k', size=20)
plt.yticks(color='k', size=20)
plt.show(True)
```
